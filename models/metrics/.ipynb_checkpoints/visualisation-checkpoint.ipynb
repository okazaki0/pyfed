```python
#import the data 
test_loss = pd.read_csv('/Users/AvyaTiK/Desktop/pyfed/Results/Testing_loss_mnist_2020-07-02_20.19.31.820075.csv', delimiter=',')
test_accuracy = pd.read_csv('/Users/AvyaTiK/Desktop/pyfed/Results/Testing_accuracy_mnist_2020-07-02_20.19.31.820075.csv',delimiter=',')
training_loss = pd.read_csv('/Users/AvyaTiK/Desktop/pyfed/Results/Training_loss_mnist_2020-07-02_20.19.31.820075.csv',delimiter=',')
```

```python
#plot the loss test
rounds = range(len(test_loss))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Test loss")
loss.plot(rounds,test_loss.values)
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
#plot the accuracy test
rounds = range(len(test_accuracy))
fig, acc = plt.subplots(figsize=(10,5))
plt.title("Test accuracy")
acc.plot(rounds,test_accuracy.values)
plt.xlabel('Rounds')
plt.ylabel('accuracy')
acc.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
#prepare the training loss data
data = prepare_data(training_loss)
```

```python
#calculate the micro loss
loss_mic = micro_loss(data)
```

```python
#plot the micro loss
rounds = range(len(loss_mic))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Micro Average Training Loss")
loss.plot(rounds,loss_mic)
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
rounds = range(len(test_loss))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Micro Average Training Loss And Test Loss")
loss.plot(rounds,loss_mic,label="Average train loss")
loss.plot(rounds,test_loss,label="Test loss")
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend(loc='upper right')
plt.show()
```
