```python
import numpy as np
import random
def f(n,m):
    data = [random.randint(1, n) for i in range(100)]
    data = np.array(data)
    while True:
        if np.sum(data)<=m :
            rest = m-np.sum(data)
            #random index
            idx = random.randint(1, 100)
            data[idx] = data[idx]+rest
            print('finich this shit, rest')
            break
        idx = np.argmax(data)
        data[idx] = random.randint(1, n)
    return data
```
