```python
#import the data 
#iid
test_loss = pd.read_csv('iid/Testing_loss_shakespeare_2020-07-09_05.23.29.827582.csv', delimiter=',')
test_accuracy = pd.read_csv('iid/Testing_accuracy_shakespeare_2020-07-09_05.23.29.827582.csv',delimiter=',')
training_loss = pd.read_csv('iid/Training_loss_shakespeare_2020-07-09_05.23.29.827582.csv',delimiter=',')
```

```python
#plot the loss test
rounds = range(len(test_loss))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Test loss")
loss.plot(rounds,test_loss.values)
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
#plot the loss test
rounds = range(len(test_accuracy))
fig, acc = plt.subplots(figsize=(10,5))
plt.title("Test accuracy")
acc.plot(rounds,test_accuracy.values)
plt.xlabel('Rounds')
plt.ylabel('accuracy')
acc.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
data = prepare_data(training_loss)
```

```python
#plot the micro loss
fig = plt.figure(figsize=(10,10))
for i,d in enumerate(data):
    plt.title("Micro Average Training Loss")
    plt.plot([data for data in d.values()], label="Round "+ str(i))
    plt.ylabel('loss')
    
plt.legend(loc='upper right')
plt.show()

```

```python
#calculate the micro loss
loss_mic = micro_loss(data)
```

```python
#plot the micro loss
rounds = range(len(loss_mic))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Micro Average Training Loss")
loss.plot(rounds,loss_mic)
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
#plot the micro loss
rounds = range(len(test_loss))
fig, loss = plt.subplots(figsize=(10,5))
plt.title("Micro Average Training Loss And Test Loss")
loss.plot(rounds,loss_mic,label="Average train loss")
loss.plot(rounds,test_loss,label="Test loss")
plt.xlabel('Rounds')
plt.ylabel('loss')
loss.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend(loc='upper right')
plt.show()
```
